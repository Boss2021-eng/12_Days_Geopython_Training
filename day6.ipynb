### Geometric Operations with Geopandas

#### Dissolve

```python
world_gdf = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world_gdf_dissolved = world_gdf.dissolve(by='continent')
world_gdf.explore()
```

#### Clip

```python
world_cities_gdf = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
#world_cities_gdf.plot()

# getting the african boundaries
africa = world_gdf[world_gdf['continent'] == 'Africa']

# performing the clip operation
africa_cities = world_cities_gdf.clip(africa)

africa_cities.plot()
```

### Buffer

```python
world_cities_gdf = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
#world_cities_gdf.plot()

# getting the african boundaries
africa = world_gdf[
    (world_gdf['continent'] == 'Africa') &
    (world_gdf['name'].str.contains("N"))
    ]

# performing the clip operation
africa_cities = world_cities_gdf.clip(africa)

africa_buffer = africa_cities.buffer(1)

africa_buffer.plot(figsize=(9,6), alpha=0.5)
```

### Plotting with Geopandas

```python
world_cities_gdf = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
#world_cities_gdf.plot()

# getting the african boundaries
africa = world_gdf[
    (world_gdf['continent'] == 'Africa') 
    ]

# performing the clip operation
africa_cities = world_cities_gdf.clip(africa)

base = africa.plot(
    figsize= (16,24),
    column='name',
    legend=True,
    cmap='Paired',
    legend_kwds= {
        'title': "African countries",
        'shadow': True,
        'ncols': 3,
        'loc': 'lower left'
    },
    zorder=1
)

# # Add buffer
# africa_buffer = africa_cities.buffer(1)
# africa_buffer.plot(ax=base, alpha=0.5, zorder=3 )

# Add africa cities layer
africa_cities.plot(ax=base, color='black' , zorder=4)

africa.boundary.plot(ax=base, zorder=2)
# base.set_axis_off()
base.grid(animated=True)


base.set_title('CITIES IN AFRICA', fontsize=25, fontweight=3)
# africa_buffer.plot(figsize=(9,6), alpha=0.5)

# Label the cities
for idx, row in africa_cities.iterrows():
    base.text(row.geometry.x, row.geometry.y, row['name'], fontsize=8, color='black', fontweight=2,
            ha='right', va='center', zorder=6)

# Label the countries
# for idx, row in africa.iterrows():
#     # Using the centroid of each country for labeling
#     base.text(row.geometry.centroid.x, row.geometry.centroid.y, row['name'], fontsize=9, color='blue',
#             ha='center', va='center', zorder=6, fontweight='bold')
```
