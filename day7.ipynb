### Introduction to Rasterio

```python
import rasterio as rio
```

```python
with rio.open('.\data\output_SRTMGL1.tif') as src:
    bands = src.count
    print(f"Number of bands : {bands}")

    # Read  a specific band in a numpy array
    band = src.read(1)

    print(band)
```

```python
from rasterio.plot import show

show(band)
```
